#### 1. List all the comedy movies.

```python
query1 = """ SELECT m.title as movie
FROM Movie m
JOIN M_Genre mg ON m.MID = mg.MID
JOIN Genre g ON mg.GID = g.GID
WHERE g.Name LIKE '%comedy%'"""

res = pd.read_sql_query(query1,conn)
res
```

#### 22. Find the total number of movies made in those years in which at least one English movie was released in that year. This question is used to find out the percentage of English movies made in each year in the following question.

```python
query22 = """ 
SELECT English_Movies_Table.Year, Total_Movies_Table.Total_Movies, English_Movies_Table.English_language_movies 
FROM

(SELECT Year, SUM(English_language_movies) AS English_language_movies FROM
(SELECT CAST(SUBSTR(m.year,-4) AS Integer) as Year, COUNT(m.MID) as English_language_movies
FROM
Movie m 
JOIN M_Language ml ON m.MID = ml.MID
JOIN Language l ON ml.LAID = l.LAID
WHERE l.LAID = 0
GROUP BY Year
ORDER BY Year DESC)
GROUP BY Year
ORDER BY Year DESC) English_Movies_Table

LEFT JOIN

(SELECT CAST(SUBSTR(m.year,-4) AS Integer) as Year, COUNT(m.MID) as Total_movies
FROM
Movie m 
GROUP BY CAST(SUBSTR(m.year,-4) AS Integer)
ORDER BY Year DESC) Total_Movies_Table

ON
English_Movies_Table.Year = Total_Movies_Table.Year
"""

res = pd.read_sql_query(query22,conn)
res
```

#### 23. Find the percentage of English movies made in each year.

```python
query23 = """
SELECT Year, Total_Movies, English_language_Movies, 
((English_language_movies * 100.00) / Total_Movies) as Percentage_of_English_Movies

FROM

(SELECT English_Movies_Table.Year, Total_Movies_Table.Total_Movies, English_Movies_Table.English_language_movies 
FROM
(SELECT Year, SUM(English_language_movies) AS English_language_movies FROM
(SELECT CAST(SUBSTR(m.year,-4) AS Integer) as Year, COUNT(m.MID) as English_language_movies
FROM
Movie m 
JOIN M_Language ml ON m.MID = ml.MID
JOIN Language l ON ml.LAID = l.LAID
WHERE l.LAID = 0
GROUP BY Year
ORDER BY Year DESC)
GROUP BY Year
ORDER BY Year DESC) English_Movies_Table
LEFT JOIN
(SELECT CAST(SUBSTR(m.year,-4) AS Integer) as Year, COUNT(m.MID) as Total_movies
FROM
Movie m 
GROUP BY CAST(SUBSTR(m.year,-4) AS Integer)
ORDER BY Year DESC) Total_Movies_Table
ON
English_Movies_Table.Year = Total_Movies_Table.Year)

"""

res = pd.read_sql_query(query23,conn)
res
```
